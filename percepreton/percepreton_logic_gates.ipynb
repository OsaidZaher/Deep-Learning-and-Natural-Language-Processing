```python
import matplotlib.pyplot as plt
from sklearn.linear_model import Perceptron
import numpy as np
from itertools import product
```

```python
data = [[0, 0], [1, 0], [1, 1], [0, 1]]
labels = [0, 0, 1, 0]

data_xor = [[0, 0], [1, 0], [1, 1], [0, 1]]
labels_xor = [0, 1, 0, 1]

data_or = [[0, 0], [1, 0], [1, 1], [0, 1]]
labels_or = [0, 1, 1, 1]
```

```python
plt.scatter([point[0] for point in data], [point[1] for point in data], c=labels, cmap='viridis')
plt.colorbar(label="Label")
plt.xlabel("X1")
plt.ylabel("X2")
plt.title("AND Gate Data Points with Labels")
plt.show()
```

```python
classifier = Perceptron(max_iter=40, random_state=22)
classifier.fit(data, labels)
res = classifier.score(data, labels)
print(f"Score on AND data: {res}")
```

```python
classifier.fit(data_xor, labels_xor)
res_xor = classifier.score(data_xor, labels_xor)
print(f"Score on XOR data: {res_xor}")
```

```python
classifier.fit(data_or, labels_or)
res_or = classifier.score(data_or, labels_or)
print(f"Score on OR data: {res_or}")
```

```python
print(classifier.decision_function([[0, 0], [1, 1], [0.5, 0.5]]))
```

```python
x_values = np.linspace(0, 1, 100)
y_values = np.linspace(0, 1, 100)
point_grid = np.array(list(product(x_values, y_values)))
distances = classifier.decision_function(point_grid)  # Compute distances from decision boundary
distances_matrix = distances.reshape(100, 100) 
```

```python
plt.pcolormesh(x_values, y_values, distances_matrix, shading='auto', cmap='coolwarm')
plt.colorbar(label="Distance from Decision Boundary")
plt.xlabel("X1")
plt.ylabel("X2")
plt.title("Decision Boundary Heatmap")
plt.show()
```
